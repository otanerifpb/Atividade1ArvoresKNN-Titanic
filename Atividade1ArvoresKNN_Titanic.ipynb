Passo 01 - Carregar a base TITANIC do seaborn

```python
from seaborn import load_dataset

df = load_dataset('titanic')

df
```

Passo 02 - Observar a relação dos atributos "pclass" com "class"

```python
# Os valores de 'pclass' e 'class', são idênticos, porem um é numérico e outro é categórico, vou eliminar a coluna 'class' por ser categórica
df = df.drop('class', axis = 1)

df
```

Passo 03 -  Observar a quantidade de NaN existente no atributo "deck"

```python
import pandas as pd

df = pd.DataFrame(data = df)

coluna = df['deck']
cont_nan = coluna.isna().sum()

# Devido a quantidade alta de NaN na coluna 'deck', será melhor remover a mesma
df = df.drop('deck', axis = 1)

print()
print('Quantidade de Linhas que contem NaN')
print("Total NaN: " + str(cont_nan))
print()
print('df com a coluna deck removida')
df
```

Passo 04 - Remover as linhas que teham valores NaN

```python
df = df.dropna()

df
```

Passo 05 - Transformar todos os atributos categóricos usando One-Hot Encoder

```python
from sklearn import preprocessing
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = make_column_transformer((OneHotEncoder(), ['sex','embarked','who', 'adult_male', 'embark_town', 'alive', 'alone']), remainder='passthrough')

df = one_hot_encoder.fit_transform(df)
df = pd.DataFrame(data = df)

df
```

Passo 06 - Coluna target é a "pclass" ou "survived"

```python
# A escolha para a Target(y) foi a coluna 'pclass', que é a coluna 16
y = df[18]

# Para as Features(X), são todas as demais colunas do df - y
X = df.drop(18, axis=1)

print(y)
print()
print(X)
```

Passo 07 - Separar os dados de treinamento e testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = None, stratify = y)
```

Passo 08 - Treinar 2 árvores de decisão com características diferentes

```python
from sklearn.tree import DecisionTreeClassifier

# Árvore de Decisão com entropy
treino1 = DecisionTreeClassifier(criterion = "entropy")
# Árvore de Decisão com profundidade limitada
treino2 = DecisionTreeClassifier(max_depth = 3)

# Treinameto dos dados com Árvore de Decesão
treino1.fit(X_train, y_train)
treino2.fit(X_train, y_train)
```

Passo 09 - Treinar 2 kNN com características diferentes

```python
from sklearn.neighbors import KNeighborsClassifier

# K-NN com 7 vizinhos, metrica 'euclidoen'
treino3 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean', algorithm = 'brute')
# K-NN com 5 vizinhos, metrica 'manhattan'
treino4 = KNeighborsClassifier(n_neighbors = 7, metric = 'manhattan', algorithm = 'brute')

# Treinameto dos dados com Árvore de Decesão
treino3.fit(X_train, y_train)
treino4.fit(X_train, y_train)
```

Passo 10 - Exibir acurácia dos testes

```python
from sklearn import metrics

print('*'*13 + 'Acurácia Árvore Decisão' + '*'*15)
result1 = treino1.predict(X_test)
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Árvore Decisão Profundidade livre: " + ' '*13 + "{}%".format(show1))

result2 = treino2.predict(X_test)
acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Árvore Decisão Profundidade 3: " + ' '*17 + "{}%".format(show1))

print()
print('*'*19 + 'Acurácia K-NN' + '*'*19)
result3 = treino3.predict(X_test)
acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Protocolo K-NN 7 Neighbord Metric 'eucledean': " + ' ' + "{}%".format(show3))

result4 = treino4.predict(X_test)
acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Protocolo K-NN 7 Neighbord  Metric 'manhattan': " + "{}%".format(show3))
```
